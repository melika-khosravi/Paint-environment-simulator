```python
import matplotlib.pyplot as plt

import numpy as np
import cv2 

drawing = False # true if mouse is pressed

# mouse callback function
def brush(event,x,y,flags,param):
    global drawing,color
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            spots = np.array([(x+60,y+30),(x-60,y+10) , (x+50,y-30)])
            cv2.drawContours(img, [spots], 0, color, -1)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
            
img = np.ones((700,700,3), np.uint8)*255
cv2.namedWindow('image')
cv2.setMouseCallback('image',brush)
color=(0,0,0)
while True:
    cv2.imshow('image',img)
    key=cv2.waitKey(1) & 0xFF
    if  key== 27:
        break
    elif key==ord('b'):
        color=(255,0,0)
    elif key==ord('g'):
        color=(0,255,0)
    elif key==ord('r'):
        color=(0,0,255)
    elif key==ord('p'):
        color=(255,0,255)
    elif key==ord('y'):
        color=(127,255,255)
    elif key==ord('g'):
        color=(127,127,127)
cv2.destroyAllWindows()
plt.imshow(img[...,::-1])
```
